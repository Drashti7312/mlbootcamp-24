```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame({'Category' : ['A','A','B','B','C','C'],'Value' : [1,2,3,4,5,6],'Value2' : [10,20,30,40,50,60]})
```

```python
data
```

```python
sns.scatterplot(data,x='Category',y='Value')
plt.show()
```

```python
sns.lineplot(data,x='Category',y='Value')
plt.show()
```

```python
plt.figure(figsize=(5,5))

data.plot(kind='bar',stacked=True,color=['red','blue'])

plt.show()
```

```python
sns.histplot(data,x='Category',y='Value')
plt.show()
```

```python
sns.violinplot(data,x='Category',y='Value')
plt.show()
```
