# line Plot

```python
x = [1,2,3,4,5]
y = [10,5,2,6,3]


plt.plot(x,y)
plt.title('X vs Y Line Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
x,y
```

```python
plt.bar(x,y,color = 'blue',edgecolor = 'red',label = 'Simple Bar')
plt.title('X vs Y Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show()
```
