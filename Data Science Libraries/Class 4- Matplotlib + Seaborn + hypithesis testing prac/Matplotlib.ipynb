```python
titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# Combining multiple plots
fig,ax = plt.subplots(3, 2, figsize=(15,15),dpi=500)

# Age distribution
ax[0, 0].hist(titanic['Age'].dropna(), bins=30, color='c')
ax[0, 0].set_title('Age Distribution')
ax[0, 0].set_xlabel('Age')
ax[0, 0].set_ylabel('Frequency')

# Fare distribution
ax[0, 1].hist(titanic['Fare'], bins=30, color='orange')
ax[0, 1].set_title('Fare Distribution')
ax[0, 1].set_xlabel('Fare')
ax[0, 1].set_ylabel('Frequency')

# Survival distribution
survived_counts = titanic['Survived'].value_counts()
ax[1, 0].pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
ax[1, 0].set_title('Survival Distribution')

# Passenger class distribution
pclass_counts = titanic['Pclass'].value_counts()
ax[1, 1].bar(pclass_counts.index, pclass_counts.values, color=['blue', 'yellow', 'green'])
ax[1, 1].set_title('Passenger Class Distribution')
ax[1, 1].set_xlabel('Class')
ax[1, 1].set_ylabel('Number of Passengers')

# Age vs Fare scatter plot
ax[2, 0].scatter(titanic['Age'], titanic['Fare'], alpha=0.5, color='purple')
ax[2, 0].set_title('Age vs Fare')
ax[2, 0].set_xlabel('Age')
ax[2, 0].set_ylabel('Fare')

# Age box plot
ax[2, 1].boxplot(titanic['Age'].dropna())
ax[2, 1].set_title('Age Boxplot')
ax[2, 1].set_ylabel('Age')

plt.tight_layout()
plt.show()
```

    -   `plt.xlim()` / `plt.ylim()`: Set the limits for the x-axis and y-axis, respectively.

    -   `plt.xticks()` / `plt.yticks()`: Set the locations and labels of ticks on the x-axis and y-axis, respectively.

    -   `plt.grid()`: Enable or disable grid lines on the plot.

    -   `plt.legend()`: Add a legend to the plot to label different lines or markers.

    -   `plt.tight_layout()`: Automatically adjust subplot parameters to give specified padding.

    -   `plt.savefig()`: Save the plot to a file (e.g., PNG, PDF, SVG).

    -   `plt.subplots_adjust()`: Adjust the spacing between subplots.

    -   `plt.axhline()` / `plt.axvline()`: Add horizontal or vertical lines to the plot.

    -   `plt.text()`: Add text to the plot at specified coordinates.

    -   `plt.annotate()`: Add annotations with arrows to points on the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Sin')
plt.plot(x, y2, label='Cos')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Plot Title')
plt.xlim(0, 10)  # Set x-axis limits
plt.ylim(-1, 1)  # Set y-axis limits
plt.xticks(np.arange(0, 11, 2))  # Set custom tick locations on x-axis
plt.yticks([-1, 0, 1], ['Min', 'Mid', 'Max'])  # Set custom tick labels on y-axis
plt.grid(True)  # Show grid lines
plt.legend()  # Show legend
plt.tight_layout()  # Adjust layout
plt.savefig('plot.png')  # Save plot to file
plt.show()
```

# Legend

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,label='Sin of x')
plt.plot(x,z,label='Cos of x')


plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Sin and Cos Graph')

plt.grid(True)

plt.legend()
plt.show()

```
