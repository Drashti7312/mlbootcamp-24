# Introduction to Conditional Statements in Python

Conditional statements are fundamental in programming as they allow the program to make decisions based on certain conditions. They enable the execution of different blocks of code depending on whether a condition is true or false. Let's explore the four main types of conditional statements in Python:

#### 1. 'if' Statement:

The `if` statement executes a block of code if the specified condition is true. It is often used for simple decision-making.

```python
# Example:
num = 10
if num > 0:
    print("Number is positive")
```

#### 2. 'if-else' Statement:

The `if-else` statement executes one block of code if the condition is true and another block if the condition is false. It allows for two possible outcomes.

```python
# Example:
num = -5
if num > 0:
    print("Number is positive")
else:
    print("Number is non-positive")
```

#### 3. Nested 'if-elif' Statement:

The nested `if-elif` statement allows for multiple conditions to be checked sequentially. Each condition is checked only if the previous conditions are false. It is useful for handling multiple scenarios.

```python
# Example:
num = 0
if num > 0:
    print("Number is positive")
elif num < 0:
    print("Number is negative")
else:
    print("Number is zero")
```

#### 4. Nested 'if' Statement:

Nested `if` statements are `if` statements within other `if` or `else` statements. They can be used to check for additional conditions based on the result of an outer condition. It allows for more complex decision-making.

```python
# Example:
num = 15
if num > 0:
    if num % 2 == 0:
        print("Number is positive and even")
    else:
        print("Number is positive and odd")
else:
    print("Number is non-positive")
```

### Conclusion:

Conditional statements are essential for controlling the flow of a program and implementing logic. They provide flexibility by allowing programs to make decisions dynamically based on different conditions. By mastering conditional statements, programmers can write more efficient and responsive code.
