# Simple Linear Regression Example

In this example, we'll demonstrate how to perform simple linear regression using scikit-learn.
We'll create a dataset with two variables: study hours and the score of the student.
Then, we'll use linear regression to find the best-fit line for the data.
Finally, we'll visualize the data points and the regression line to understand the intuition behind linear regression.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Define the dataset
study_hours = np.array([3, 5, 2, 7, 4, 6, 1, 3, 5, 4]).reshape(-1, 1) # X 
scores = np.array([75, 82, 69, 88, 78, 85, 62, 72, 80, 76]) # y
```

```python
plt.scatter(study_hours,scores)
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```

```python
# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(study_hours, scores)
```

```python
# Get the slope and intercept of the regression line
slope = model.coef_[0]
intercept = model.intercept_
slope, intercept
```

```python
slope * 8 + intercept
```

```python
model.predict(np.array(5).reshape(1,-1))
```

```python
# Visualize the data points and regression line
plt.scatter(study_hours, scores, color='blue', label='Data Points')



plt.plot(study_hours, model.predict(study_hours), color='red', label='Regression Line')



plt.title('Linear Regression Example')
plt.xlabel('Study Hours')
plt.ylabel('Score')


plt.legend()




plt.show()
```
