# RIDGE AND LASSO Practical

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
df = fetch_california_housing()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data,columns=df.feature_names)
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Price']  = df.target
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1]
y= dataset.iloc[:,-1]
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression(n_jobs=-1)
lr
```

```python
mse = cross_val_score(lr,X,y,scoring='neg_mean_squared_error',cv = 5)

mean_mse = np.mean(mse)
print(mean_mse)
```

```python
mse
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python


alpha = [1e-15,1e-10,1e-8,1e-6,1e-2,0,1,2,5,10,20,30,35,40,50,55,60,80,100]

parameters = {'alpha': alpha}
```

```python
parameters
```

```python
ridge_regresseror = GridSearchCV(ridge,parameters,scoring = 'neg_mean_squared_error',cv=5)
```

```python
ridge_regresseror.fit(X,y)
```

```python
print(ridge_regresseror)
```

```python
print(ridge_regresseror.best_params_)
```

```python
print(ridge_regresseror.best_score_)
```

Simple LR :- -0.5582901717686569

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
alpha = [1e-15,1e-10,1e-8,1e-6,1e-2,0,1,2,5,10,20,30,35,40,50,55,60,80,100]

parameters = {'alpha': alpha}
```

```python
lasso_regressor = GridSearchCV(lasso,parameters,scoring = 'neg_mean_squared_error',cv=5)
```

```python
lasso_regressor.fit(X,y)
```

```python
lasso_regressor.best_params_
```

```python
lasso_regressor.best_score_
```

```python
-0.5579339846725311
Simple LR :- -0.5582901717686569
```
