# K-Nearest Neighbors (KNN) Classification

In this notebook, we'll implement the K-Nearest Neighbors (KNN) algorithm. We'll go through each step in detail, explaining the concepts and code to ensure a thorough understanding.

## Importing Libraries

First, we'll import the necessary libraries.

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline
```

## Loading the Dataset

We'll use the Iris dataset for our KNN implementation. Let's load and inspect the dataset.

```python
# Loading the Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                    columns= iris['feature_names'] + ['species'])

# Displaying the first few rows of the dataset
print(data.head())
```

## Visualizing the Dataset

Let's visualize the dataset to understand its structure. We'll create pair plots to see the relationships between different features.

```python
# Creating pair plots
sns.pairplot(data, hue='species', markers=['o', 's', 'D'])
plt.show()
```

## Splitting the Dataset

We'll split the dataset into training and testing sets to evaluate our KNN model.

```python
# Splitting the dataset into training and testing sets
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Feature Scaling

Feature scaling is essential for KNN as it is a distance-based algorithm. We'll use StandardScaler for scaling our features.

```python
# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## KNN Algorithm

The K-Nearest Neighbors algorithm is a simple, instance-based learning algorithm. It classifies a data point based on the majority class among its k-nearest neighbors.

### Choosing the value of k

Choosing the right value for k is crucial for the performance of the KNN classifier. We'll use the elbow method to select the best k.

```python
# Function to calculate the Euclidean distance between two points
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))
```

```python
# Function to get the k-nearest neighbors
def get_neighbors(X_train, y_train, test_point, k):
    distances = []
    for i in range(len(X_train)):
        dist = euclidean_distance(X_train[i], test_point)
        distances.append((y_train[i], dist))
    distances.sort(key=lambda x: x[1])
    neighbors = [distances[i][0] for i in range(k)]
    return neighbors
```

```python
# Function to make a prediction based on majority voting
def predict(X_train, y_train, test_point, k):
    neighbors = get_neighbors(X_train, y_train, test_point, k)
    most_common = Counter(neighbors).most_common(1)
    return most_common[0][0]
```

```python
# Function to calculate the accuracy of the model
def calculate_accuracy(X_train, y_train, X_test, y_test, k):
    predictions = [predict(X_train, y_train, test_point, k) for test_point in X_test]
    return accuracy_score(y_test, predictions)
```

```python
# Finding the best k using the elbow method
k_values = range(1, 11)
accuracies = [calculate_accuracy(X_train, y_train, X_test, y_test, k) for k in k_values]

# Plotting the accuracies for different k values
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('Elbow Method for Selecting k')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Printing accuracies for each k value
for i, accuracy in enumerate(accuracies, 1):
    print(f'Accuracy for k={i}: {accuracy}')
```

## Visualizing the Decision Boundaries

Let's visualize the decision boundaries for our KNN classifier.

```python
# Function to plot decision boundaries
def plot_decision_boundaries(X, y, k):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = np.array([predict(X, y, np.array([xx.ravel()[i], yy.ravel()[i]]), k) for i in range(len(xx.ravel()))])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='deep', s=50)
    plt.title(f'KNN Decision Boundaries (k={k})')
    plt.show()
```

```python
# Plotting decision boundaries for k=3
# We'll use only the first two features for visualization
plot_decision_boundaries(X_train[:, :2], y_train, k=3)
```

In this notebook, we implemented the K-Nearest Neighbors (KNN) algorithm from scratch. We visualized the dataset, scaled the features, and used the elbow method to select the best value of k. Finally, we plotted the decision boundaries to understand how the KNN classifier works. This interactive and detailed explanation should help in understanding the KNN algorithm better.
