# Assumptions of Linear Regression

Linear regression makes several key assumptions about the relationship between the independent and dependent variables. These assumptions are important to consider when interpreting the results of a linear regression model.

### Assumption 1: Linearity

The relationship between the independent and dependent variables should be linear. This means that the change in the dependent variable is proportional to the change in the independent variable.

Mathematically, it can be represented as:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n + \epsilon $$

Where:
- $y$ is the dependent variable
- $x_1, x_2, ..., x_n$ are the independent variables
- $\beta_0, \beta_1, \beta_2, ..., \beta_n$ are the coefficients
- $\epsilon$ is the error term

Here's an example of a linear relationship between study hours and exam scores:

```python
# Generate synthetic data for illustration

```

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
study_hours = np.random.randint(1, 10, 20)
exam_scores = 2 * study_hours + np.random.normal(0, 2, 20)

# Plotting the data
plt.scatter(study_hours, exam_scores)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Linear Relationship')
plt.show()
````

### Assumption 2: Independence

The observations should be independent of each other. In other words, the value of one observation should not be influenced by the value of another observation.

For example, in a study examining the relationship between study hours, students' sleep hours, and exam scores, each student's exam score should be independent of other students' scores.

### Assumption 3: Multicollinearity

The independent variables should not be highly correlated with each other. High correlation between independent variables can lead to multicollinearity, which can cause issues with the model's interpretability and stability.

For example, in a study examining the relationship between study hours, students' sleep hours, and exam scores, if study hours and sleep hours are highly correlated (students who study more tend to sleep less), it can lead to multicollinearity.

### Assumption 4: Homoscedasticity

The variance of the residuals (the differences between the observed and predicted values) should be constant across all levels of the independent variable.

In simpler terms, the spread of the residuals should be consistent along the range of predicted values.



````python

# Generate synthetic data for illustration
np.random.seed(0)
predicted_values = np.random.randint(50, 100, 20)
residuals = np.random.normal(0, 10, 20)

# Plotting the residuals
plt.scatter(predicted_values, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Homoscedasticity')
plt.axhline(y=0, color='r', linestyle='--')  # Adding horizontal line at y=0
plt.show()
````

### Assumption 5: Feature Scaling Required?

Linear regression assumes that the features are on a similar scale. If the features are on vastly different scales, it may cause convergence issues and affect the performance of the model.

For example, in a study examining the relationship between study hours (measured in hours) and students' income (measured in dollars), the features are on vastly different scales. In this case, feature scaling (e.g., normalization or standardization) may be required.

### Assumption 6: Heteroscedasticity

Heteroscedasticity occurs when the variance of the residuals is not constant across all levels of the independent variable. In other words, the spread of the residuals varies along the range of predicted values.


````python

# Generate synthetic data for illustration
np.random.seed(0)
predicted_values = np.random.randint(50, 100, 20)
residuals = np.random.normal(0, 10, 20) + predicted_values * 0.1

# Plotting the residuals
plt.scatter(predicted_values, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Heteroscedasticity')
plt.axhline(y=0, color='r', linestyle='--')  # Adding horizontal line at y=0
plt.show()
````
